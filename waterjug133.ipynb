```python
  # BFS PROGRAM
from collections import deque

def get_next_states(x, y, a, b):
    states = []

    # Fill Jug A
    if x < a:
        states.append((a, y, f"Fill Jug A ({x}->{a})"))
    # Fill Jug B
    if y < b:
        states.append((x, b, f"Fill Jug B ({y}->{b})"))
    # Empty Jug A
    if x > 0:
        states.append((0, y, f"Empty Jug A ({x}->0)"))
    # Empty Jug B
    if y > 0:
        states.append((x, 0, f"Empty Jug B ({y}->0)"))
    # Pour A -> B
    if x > 0 and y < b:
        pour = min(x, b - y)
        states.append((x - pour, y + pour, f"Pour {pour} from A -> B"))
    # Pour B -> A
    if y > 0 and x < a:
        pour = min(y, a - x)
        states.append((x + pour, y - pour, f"Pour {pour} from B -> A"))

    return states


def water_jug_bfs(a, b, target):
    visited = set()
    queue = deque([((0, 0), [], [])])  # state, path, actions

    while queue:
        (x, y), path, actions = queue.popleft()

        if (x, y) in visited:
            continue
        visited.add((x, y))
        path = path + [(x, y)]

        if x == target or y == target:
            return path, actions

        for nx, ny, action in get_next_states(x, y, a, b):
            if (nx, ny) not in visited:
                queue.append(((nx, ny), path, actions + [action]))

    return None, None


# Main
a = int(input("Enter capacity of Jug A: "))
b = int(input("Enter capacity of Jug B: "))
target = int(input("Enter target amount: "))

path, actions = water_jug_bfs(a, b, target)

print("\n--- BFS Solution with Detailed Steps ---")
if path:
    for i, (state, action) in enumerate(zip(path[1:], actions), 1):
        print(f"Step {i}: {action} => Jug A = {state[0]}, Jug B = {state[1]}")
else:
    print("No solution found")
```
