```python
from collections import deque

# Check if all rooms are clean
def is_goal(rooms):
    return all(status == "Clean" for status in rooms.values())

# Generate possible actions from current state
def get_successors(state, room_names):
    location, rooms = state
    successors = []

    # Copy room status to avoid mutation
    rooms_copy = rooms.copy()

    # Action: Suck
    if rooms_copy[location] == "Dirty":
        rooms_copy[location] = "Clean"
        successors.append(("Suck", (location, rooms_copy)))

    # Move actions: go to any other room
    for room in room_names:
        if room != location:
            successors.append((f"Move to {room}", (room, rooms_copy)))

    return successors

# BFS to find sequence of actions
def bfs(initial_state, room_names):
    queue = deque()
    queue.append((initial_state, []))  # (state, path)
    visited = set()

    while queue:
        current_state, path = queue.popleft()
        location, rooms = current_state

        # Serialize state for visited check
        state_key = (location, tuple(sorted(rooms.items())))
        if state_key in visited:
            continue
        visited.add(state_key)

        if is_goal(rooms):
            return path  # sequence of actions

        for action, next_state in get_successors(current_state, room_names):
            queue.append((next_state, path + [action]))

    return None  # No solution found

# --- Input ---
num_rooms = int(input("Enter number of rooms: "))
room_names = []
rooms = {}

for i in range(num_rooms):
    room_name = input(f"Enter name for Room {i+1}: ").strip()
    room_status = input(f"Enter status of {room_name} (Clean/Dirty): ").capitalize()
    room_names.append(room_name)
    rooms[room_name] = room_status

start_location = input(f"Enter starting location of agent (choose from {room_names}): ").strip()

initial_state = (start_location, rooms)

# --- Run BFS ---
solution = bfs(initial_state, room_names)

# --- Simulate steps ---
if solution:
    current_location = start_location
    rooms_copy = rooms.copy()
    print("\nInitial State:", rooms_copy)

    for step_num, action in enumerate(solution, 1):
        print(f"\nStep {step_num}:")
        print("Agent Location:", current_location)
        print("Room Status:", rooms_copy[current_location])
        print("Action:", action)

        # Execute action
        if action == "Suck":
            rooms_copy[current_location] = "Clean"
        elif action.startswith("Move"):
            current_location = action.split()[-1]

        print("Updated Rooms:", rooms_copy)

    print("\nFinal State:", rooms_copy)
else:
    print("No solution found.")
```
