```python
import itertools
import pandas as pd
```

```python
# P(Rain)
P_R = {
    True: 0.2,
    False: 0.8
}

# P(Sprinkler | Rain)
P_S_given_R = {
    True:  {True: 0.01, False: 0.99},   # Rain = True
    False: {True: 0.4,  False: 0.6}     # Rain = False
}

# P(WetGrass | Sprinkler, Rain)
P_W_given_SR = {
    (True, True):   {True: 0.99, False: 0.01},
    (True, False):  {True: 0.90, False: 0.10},
    (False, True):  {True: 0.80, False: 0.20},
    (False, False): {True: 0.0,  False: 1.0}
}
```

```python
def joint_probability(r, s, w):
    return (
        P_R[r] *
        P_S_given_R[r][s] *
        P_W_given_SR[(s, r)][w]
    )
```

```python
variables = [True, False]

data = []

for r, s, w in itertools.product(variables, variables, variables):
    prob = joint_probability(r, s, w)
    data.append([r, s, w, round(prob, 5)])

df = pd.DataFrame(data, columns=["Rain", "Sprinkler", "WetGrass", "Probability"])
df
```

```python
def marginal_W(w_value):
    total = 0
    for r, s in itertools.product([True, False], [True, False]):
        total += joint_probability(r, s, w_value)
    return total

print("P(WetGrass=True)  =", round(marginal_W(True), 5))
print("P(WetGrass=False) =", round(marginal_W(False), 5))
```

```python
def conditional_W_given_R(w_value, r_value):
    numerator = 0
    denominator = 0
    
    for s in [True, False]:
        numerator += joint_probability(r_value, s, w_value)
        denominator += (
            joint_probability(r_value, s, True) +
            joint_probability(r_value, s, False)
        )
    
    return numerator / denominator

print("P(WetGrass=True | Rain=True) =", round(conditional_W_given_R(True, True), 5))
```

```python
def posterior_R_given_W(r_value, w_value):
    numerator = 0
    denominator = 0
   
    for r in [True, False]:
        for s in [True, False]:
            jp = joint_probability(r, s, w_value)
           
            if r == r_value:
                numerator += jp
           
            denominator += jp
   
    return numerator / denominator

# Compute P(R=True | W=True)
print("P(R=True | W=True) =", round(posterior_R_given_W(True, True), 5))

# Compute P(R=False | W=True)
print("P(R=False | W=True) =", round(posterior_R_given_W(False, True), 5))
```
